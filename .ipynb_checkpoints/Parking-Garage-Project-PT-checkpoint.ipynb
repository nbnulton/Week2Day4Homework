```python
# Parking Garage:

# dictionary {ticket, space}
#from IPython.display import clear_output

#class Screen():

#    def __init__(self, parking):
#        self.parking = parking
        
#    def run_program(self):
#        while True:
#            response = input("What do you want to do: park, pay, or leave?")
#            if response.lower == "park":
#                self.parking.park()
#            if response.lower == "pay":
#                self.parking.pay()
#            if response.lower == "leave":
#                self.parking.leave()
    

#class Parking():

#    def __init__(self, tickets, parking_spaces, currentTicket):
#        self.tickets = [1,2,3,4,5]
#        self.parking_spaces = [1,2,3,4,5]
#        self.current_ticket = {}
        
#    def park(self, tickets, parking_spaces, currentTicket:
#        print("Parking: Printing ticket/space #: ", {tickets[0]})
#        self.tickets[0] = self.current_ticket {self.tickets[0],""}
#        self.parking_spaces = self.current_ticket {"", self.parking_spaces[0]}
#        self.tickets[0] = self.tickets.remove()
#        self.parking_spaces[0] = self.parking_spaces.remove()
#        self.tickets = self.tickets.sort()
#        self.parking_spaces = self.parking_spaces.sort()

#    def pay(self, current_ticket):
#        ticket_number = input("Paying: Please enter ticket/space #: ")
#        if ticket_number in self.current_ticket:
#            print(f"", {ticket_number}, " - Processing Payment... Payment Complete. You have 15 minutes to leave.")
#        else:
#            input("Invalid ticket/space #. Please re-enter correct ticket/space #: ")
    
#    def leave(self, tickets, parking_spaces):
#        ticket_number = input("Leaving: Please enter ticket/space #: ")
#        if ticket_number in self.current_ticket:
#        print("Thank you, have a nice day!")
#            self.current_ticket = self.current_ticket.remove(ticket_number)
#            self.tickets = tickets.append(ticket_number)
#            parking_spaces = parking_spaces.append(ticket_number)
#            self.tickets = self.tickets.sort()
#            self.parking_spaces = self.parking_spaces.sort()


#screen.run_program()
#-----------------------------------------------

import os

class ParkingGarage():
                                
    def __init__(self, number_of_spaces):    #number of spaces = 5
        self.tickets = 0    #not tickets have been sold when garage is created
        self.ticket_status = {}    #key is a ticket #, value is True or False for whether the ticket has been paid or not
        self.parking_spaces = [False for x in range(number_of_spaces)]    #list that manages the 5 parking spaces, False space is empty, True someone is parked, list could show False five times
        
        
    def take_ticket(self):
        #creates new ticket for the user if the lot is not full
        if self.__is_full():
            return "Lot is full"   #can't park
        else:
            self.tickets += 1   #lot isn't full, increment the # of tickets by 1, becomes ticket number
            self.ticket_status[self.tickets] = False    #adds ticket to dictionary, and False for ticket isn't paid
            self.__take_parking_space()   #occupy parking space, take space
            return f"Thank you for parking, your ticket number is {self.tickets}. Please pay before leaving garage"
        
    def pay_for_parking(self, ticket_number):   #pass in the ticket number
        if ticket_number not in self.ticket_status:  #checks if passed in ticket number is not in the dictionary
            return("Not a valid ticket number")
        elif self.ticket_status[ticket_number] == True:  #if ticket number is in status dictionary and set to True
            return "Ticket is already paid"
        else:
            self.ticket_status[ticket_number] = True  #if none of those conditions are met, then the number must be in the status dictionary, set the False to True
            return "Thank you for paying, you have 15:00 minutes to leave the garagge"
        
    def leave_garage(self, ticket_number):  #pass in the ticket number
        if ticket_number not in self.ticket_status:  #checks if pass in ticket number is not in the dictionary
            return("Not a valid ticket number")
        elif self.ticket_status[ticket_number] == False:  #checks if status is False (not paid)
            return "Please pay before leaving"
        else:
            self.__leave_parking_space()  #if none of those conditions are met, then the number must be in the status dictionary, set the False to True
            return "Thank you for your business. Have a great day!"
        
        
#Private Methods

    def __is_full(self):
        #check to see if any spaces have a value of False, otherwise the lost is full
        for space in self.parking_spaces:  #goes through list of parking spaces
            if space == False:     #if there is a False (unoccupied)
                return False     #returns False, lot is not full
        return True        #if not False, lot is full
    
    def __take_parking_space(self):
        #find the first space with a value of False (empty), set it to True (occupied)
        for index, space in enumerate (self.parking_spaces):   #looping through list of parking spaces
            if space == False:   #finding the first space that's False (empty)
                self.parking_spaces[index] = True  #setting it to True (occupied)
                break    #returns to other method
                
    def __leave_parking_space(self):
        #find the first space with a value of True (False), set it to True (empty)
        for index, space in enumberate(self.parking_spaces):   #look through list of parking spaces
            if space == True:  #find the first True
                self.parking_spaces[index] = False  #set it to false
                break
                

                
                
class UI():
    
    def __init__(self,garage):
        self.garage = garage  #when we instantiate this class, we pass it in a garage, so the UI object has a garage attribute
                              #  and that garage attribute is an object of the class ParkingGarage
        
#Public Methods

    def start(self):
        #starts the parking garage user interface
        while True:
            self.clear_terminal()
            print("Welcome to our parking garage")
            response = input("\n 1 park \n 2 Pay \n 3 Leave \n 4 Exit \n Please choose an option: ")
            self.clear_terminal()
            if response == "1":
                print(self.garage.take_ticket())
                input("\nPress enter to continue")
            elif response == "2":
                self.clear_terminal()
                ticket = ("Please enter your ticket number: ")
                while not ticket.isnumeric():
                    ticket = input("Please enter a number: ")
                    print(self.garage.pay_for_parking(int(ticket)))
                    input("\nPress enter to continue ")
            elif response == "3":
                self.clear_terminal()
                ticket = ("Please enter your ticket number: ")
                while not ticket.isnumeric():
                    ticket = input("Please enter a number: ")
                    print(self.garage.leave_garage(int(ticket)))
                    input("\nPress enter to continue ")
            elif response == "4":
                self.clear_terminal()
                print("Program Exit")
                break
            elif response == "5":
                print(self.garage.tickets)
                print(self.garage.ticket_status)
                print(self.garage.parking_spaces)
                input("Press any key to continue")
            else:
                print("Please enter a valid option")
                
#Private Methods

    def clear_terminal(self):
        #clears the terminal output
        os.system('cls' if os.name=='nt' else 'clear')
        
        
#Driver code
my_garage = ParkingGarage(5)
ui = UI(my_garage)
ui.start()
        
```
